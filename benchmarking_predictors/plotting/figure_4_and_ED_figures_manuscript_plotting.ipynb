```python
#@title Build final main and supplementary benchmark figures as JPEGs with updated Fig. 4A/4B styling
from __future__ import annotations

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import spearmanr, pearsonr
from warnings import catch_warnings, simplefilter
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import colors as mcolors
from matplotlib.patches import Patch

# ============================================================
# INPUT / OUTPUT
# ============================================================

"""
Repository usage
----------------

Input required:
Use the manuscript supplementary dataset:

Supplementary Table 12:
"OpenSplice dataset with splicing variant effect predictions"

This table contains:
- variant metadata
- experimental PSI values
- model predictions at canonical splice sites

for 590,104 variants across 599 exons where all models
successfully produced predictions.

Before running:
1. Download Supplementary Table 12 from the manuscript repository.
2. Update MERGED_FILE below to point to your local copy.
3. Choose your own output directory in OUT_DIR_FINAL.
"""

# ------------------------------------------------------------
# USER INPUT FILE
# ------------------------------------------------------------

MERGED_FILE = (
    "PATH/TO/"
    "Supplementary_Table_12_OpenSplice_dataset_with_splicing_variant_effect_predictions.tsv"
)

# ------------------------------------------------------------
# USER OUTPUT DIRECTORY
# ------------------------------------------------------------

OUT_DIR_FINAL = (
    "PATH/TO/YOUR/OUTPUT_DIRECTORY"
)

os.makedirs(OUT_DIR_FINAL, exist_ok=True)

print("MERGED_FILE:")
print(MERGED_FILE)

print("\nOUT_DIR_FINAL:")
print(OUT_DIR_FINAL)
EXP_COL = "delta_psi"
EXON_COL = "ensembl_exon_id"

JPEG_DPI = 300
REPLICATE_RHO_THRESHOLD = 0.8
REPLICATE_FILTER_LABEL = f"Inter-replicate experimental Spearman R > {REPLICATE_RHO_THRESHOLD}"

# ============================================================
# SETTINGS
# ============================================================

WT_PSI_MIN = 20
WT_PSI_MAX = 80
MIN_ABS_EFFECT = 0.0

MIN_MUTS_PER_EXON_GROUP = 0
MIN_MUTS_FOR_POOLED_PANEL = 2

PANEL_COLLAPSE_KEYS = ["nt_seq"]

REGION_ORDER = ["All", "Intron up", "3'SS", "Exon", "5'SS", "Intron down"]
MUT_PLOT_ORDER = ["All", "Sub", "Del", "∆1nt", "∆3nt", "∆6nt", "∆21nt"]

MODEL_FAMILY_ORDER = [
    "SpliceTransformer",
    "AlphaGenome",
    "Pangolin",
    "SpliceAI",
]

MODEL_ORDER = [
    "SpliceTransformer genome",
    "SpliceTransformer minigene",
    "AlphaGenome genome",
    "AlphaGenome minigene",
    "Pangolin genome",
    "Pangolin minigene",
    "SpliceAI genome",
    "SpliceAI minigene",
]

PALETTE = {
    "SpliceTransformer genome":   "#CC7F86",
    "SpliceTransformer minigene": "#99000D",
    "AlphaGenome genome":         "#9BC7A6",
    "AlphaGenome minigene":       "#2E7D32",
    "Pangolin genome":            "#8FA7C7",
    "Pangolin minigene":          "#1D4E89",
    "SpliceAI genome":            "#E9C46A",
    "SpliceAI minigene":          "#D4A017",
}

PRED_SCALE_FOR_ERROR = 100.0

plt.rcParams.update({
    "pdf.fonttype": 42,
    "ps.fonttype": 42,
    "font.size": 9,
    "axes.titlesize": 9,
    "axes.labelsize": 9,
    "xtick.labelsize": 7,
    "ytick.labelsize": 7,
})

# ============================================================
# LOAD + SAFE COMPLETE-MODEL FILTER
# ============================================================
# ============================================================
# LOAD + GLOBAL SAFE UNIQUE-NT_SEQ FILTER
# ============================================================

SOURCE_DF_RAW = pd.read_csv(MERGED_FILE, sep="\t", low_memory=False)

main_score_cols = [
    "spliceai_minigene__spliceai_minigene_delta_mean",
    "spliceai_genome__spliceai_genome_delta_mean",
    "pangolin_minigene__pangolin_minigene_mean_delta_signed",
    "pangolin_genome__pangolin_genome_mean_delta_signed",
    "splicetransformer_minigene__splice_transformer_minigene_delta_mean",
    "splicetransformer_genome__splice_transformer_genomic_delta_mean",
    "alphagenome_genome__mean_delta_splice",
    "alphagenome_minigene__alphagenome_minigene_delta_mean",
]

required_cols = [
    EXP_COL,
    "nt_seq",
    "variant_id",
    "exon_id",
    "ensembl_exon_id",
    "region",
    "wt_psi",
    "psi_r1",
    "psi_r2",
    "psi_r3",
    "se_wt_psi",
    *main_score_cols,
]

missing_cols = [c for c in required_cols if c not in SOURCE_DF_RAW.columns]
if missing_cols:
    raise ValueError(f"Missing required columns:\n{missing_cols}")

if "mut" not in SOURCE_DF_RAW.columns and "mut_type" not in SOURCE_DF_RAW.columns:
    raise ValueError("Need either 'mut' or 'mut_type' column.")

SOURCE_DF = SOURCE_DF_RAW.copy()

SOURCE_DF[EXP_COL] = pd.to_numeric(SOURCE_DF[EXP_COL], errors="coerce")

print("=" * 100)
print("GLOBAL FILTER FOR ALL FIGURES")
print("=" * 100)
print("Raw rows:", len(SOURCE_DF))
print("Raw unique nt_seq:", SOURCE_DF["nt_seq"].nunique(dropna=True))

# 1. Experimental measurement must exist.
SOURCE_DF = SOURCE_DF[
    SOURCE_DF[EXP_COL].notna()
    & SOURCE_DF["nt_seq"].notna()
].copy()

print("\nAfter non-null delta_psi + nt_seq:")
print("Rows:", len(SOURCE_DF))
print("Unique nt_seq:", SOURCE_DF["nt_seq"].nunique(dropna=True))
print("Unique corrected exon_id:", SOURCE_DF["exon_id"].nunique(dropna=True))
print("Unique corrected ensembl_exon_id:", SOURCE_DF["ensembl_exon_id"].nunique(dropna=True))

# 2. All models must have predictions.
SOURCE_DF = SOURCE_DF[
    SOURCE_DF[main_score_cols].notna().all(axis=1)
].copy()

print("\nAfter complete-model prediction filter:")
print("Rows:", len(SOURCE_DF))
print("Unique nt_seq:", SOURCE_DF["nt_seq"].nunique(dropna=True))
print("Unique corrected exon_id:", SOURCE_DF["exon_id"].nunique(dropna=True))
print("Unique corrected ensembl_exon_id:", SOURCE_DF["ensembl_exon_id"].nunique(dropna=True))

# 3. Collapse to one row per unique nt_seq.
n_dup_ntseq = int(SOURCE_DF.duplicated("nt_seq").sum())
print("\nDuplicated nt_seq rows before collapse:", n_dup_ntseq)

SOURCE_DF = (
    SOURCE_DF
    .sort_values(["nt_seq", "variant_id"], kind="mergesort")
    .drop_duplicates("nt_seq", keep="first")
    .reset_index(drop=True)
    .copy()
)

SAFE_EXON_SET = set(SOURCE_DF[EXON_COL].dropna().astype(str))

print("\n" + "=" * 100)
print("FINAL SHARED FIGURE UNIVERSE")
print("=" * 100)
print("Rows / unique nt_seq:", len(SOURCE_DF))
print("Unique corrected exon_id:", SOURCE_DF["exon_id"].nunique(dropna=True))
print("Unique corrected ensembl_exon_id:", SOURCE_DF["ensembl_exon_id"].nunique(dropna=True))

print("\nCoverage per model in final universe:")
for c in main_score_cols:
    print(f"  {c}: {SOURCE_DF[c].notna().sum():,} / {len(SOURCE_DF):,}")

# ============================================================
# HELPERS
# ============================================================

def save_jpeg(fig, filename):
    out = os.path.join(OUT_DIR_FINAL, filename)
    fig.savefig(out, bbox_inches="tight", dpi=JPEG_DPI)
    print("Saved:", out)


def lighten_color(color, amount=0.62):
    c = np.array(mcolors.to_rgb(color))
    white = np.array([1.0, 1.0, 1.0])
    return tuple(c + (white - c) * amount)


def extract_variant_token(x):
    if pd.isna(x):
        return pd.NA
    s = str(x).strip()
    return s.rsplit("_", 1)[-1] if "_" in s else s


def safe_spearman_no_warning(x, y, min_n=2):
    tmp = pd.DataFrame({
        "x": pd.to_numeric(x, errors="coerce"),
        "y": pd.to_numeric(y, errors="coerce"),
    }).dropna()

    if len(tmp) < min_n:
        return np.nan
    if tmp["x"].nunique() < 2 or tmp["y"].nunique() < 2:
        return np.nan

    with catch_warnings():
        simplefilter("ignore")
        rho, _ = spearmanr(tmp["x"], tmp["y"])

    return float(rho)


def safe_pearson_stats(x, y):
    tmp = pd.DataFrame({
        "x": pd.to_numeric(x, errors="coerce"),
        "y": pd.to_numeric(y, errors="coerce"),
    }).dropna()

    n = len(tmp)
    if n < 3:
        return np.nan, np.nan, n
    if tmp["x"].nunique() < 2 or tmp["y"].nunique() < 2:
        return np.nan, np.nan, n

    r, p = pearsonr(tmp["x"], tmp["y"])
    return float(r), float(p), n


def median_iqr(x):
    x = pd.to_numeric(pd.Series(x), errors="coerce").dropna()
    if len(x) == 0:
        return np.nan, np.nan, np.nan
    return float(x.median()), float(x.quantile(0.25)), float(x.quantile(0.75))


def apply_r_panel_theme(ax):
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    ax.tick_params(
        axis="both",
        width=0.6,
        length=3,
    )


def add_mut_plot_columns(df_in):
    df = df_in.copy()

    if "variant_token" not in df.columns:
        df["variant_token"] = df["variant_id"].map(extract_variant_token).astype("string")

    if "mut" not in df.columns:
        if "mut_type" in df.columns:
            df["mut"] = df["mut_type"].astype(str)
        else:
            raise ValueError("Need either 'mut' or 'mut_type' column.")

    if "mut_type" in df.columns:
        df["mut"] = np.where(
            df["mut_type"].astype(str).eq("sub"),
            "Sub",
            df["mut"].astype(str),
        )

    df["mut_plot"] = "Other"
    df.loc[df["mut"].astype(str).eq("Sub"), "mut_plot"] = "Sub"
    df.loc[df["mut"].astype(str).str.startswith("∆"), "mut_plot"] = "Del"

    for dd in ["∆1nt", "∆3nt", "∆6nt", "∆21nt"]:
        df.loc[df["mut"].astype(str).eq(dd), "mut_plot"] = dd

    return df


def add_groupings(df_in):
    base = df_in.copy()
    parts = [base.copy()]

    tmp = base.copy()
    tmp["region"] = "All"
    parts.append(tmp)

    tmp = base.copy()
    tmp["mut_plot"] = "All"
    parts.append(tmp)

    tmp = base.copy()
    tmp["region"] = "All"
    tmp["mut_plot"] = "All"
    parts.append(tmp)

    del_mask = base["mut_plot"].astype(str).isin(["∆1nt", "∆3nt", "∆6nt", "∆21nt", "Del"])

    if del_mask.any():
        tmp = base.loc[del_mask].copy()
        tmp["mut_plot"] = "Del"
        parts.append(tmp)

        tmp = base.loc[del_mask].copy()
        tmp["region"] = "All"
        tmp["mut_plot"] = "Del"
        parts.append(tmp)

    out = pd.concat(parts, ignore_index=True)
    out = out[
        out["region"].astype(str).isin(REGION_ORDER)
        & out["mut_plot"].astype(str).isin(MUT_PLOT_ORDER)
    ].copy()

    return out


def filter_df(df, filter_mode):
    out = df.copy()
    out[EXP_COL] = pd.to_numeric(out[EXP_COL], errors="coerce")

    if "wt_psi" in out.columns:
        out["wt_psi"] = pd.to_numeric(out["wt_psi"], errors="coerce")

    needed = [
        EXP_COL,
        "region",
        "mut_plot",
        "exon_id",
        "ensembl_exon_id",
        "nt_seq",
        "variant_token",
    ]

    for c in needed:
        if c not in out.columns:
            raise ValueError(f"Required column missing: {c}")

    out = out[
        out[EXP_COL].notna()
        & out["region"].notna()
        & out["mut_plot"].notna()
        & out["nt_seq"].notna()
        & out["variant_token"].notna()
    ].copy()

    if filter_mode == "wt_psi":
        out = out[out["wt_psi"].between(WT_PSI_MIN, WT_PSI_MAX, inclusive="both")].copy()

    if filter_mode == "abs_effect":
        out = out[out[EXP_COL].abs() > MIN_ABS_EFFECT].copy()

    return out


def choose_exon_group_col(df):
    if "ensembl_exon_id" in df.columns:
        return "ensembl_exon_id"
    if "exon_id" in df.columns:
        return "exon_id"
    raise ValueError("Need ensembl_exon_id or exon_id.")


def build_model_definition_table(df):
    candidates = [
        ("SpliceTransformer genome", "SpliceTransformer", "genome",
         "splicetransformer_genome__splice_transformer_genomic_delta_mean"),
        ("SpliceTransformer minigene", "SpliceTransformer", "minigene",
         "splicetransformer_minigene__splice_transformer_minigene_delta_mean"),
        ("AlphaGenome genome", "AlphaGenome", "genome",
         "alphagenome_genome__mean_delta_splice"),
        ("AlphaGenome minigene", "AlphaGenome", "minigene",
         "alphagenome_minigene__alphagenome_minigene_delta_mean"),
        ("Pangolin genome", "Pangolin", "genome",
         "pangolin_genome__pangolin_genome_mean_delta_signed"),
        ("Pangolin minigene", "Pangolin", "minigene",
         "pangolin_minigene__pangolin_minigene_mean_delta_signed"),
        ("SpliceAI genome", "SpliceAI", "genome",
         "spliceai_genome__spliceai_genome_delta_mean"),
        ("SpliceAI minigene", "SpliceAI", "minigene",
         "spliceai_minigene__spliceai_minigene_delta_mean"),
    ]

    rows = []
    for display_name, family, mode, score_col in candidates:
        if score_col in df.columns:
            rows.append((display_name, family, mode, score_col))

    return pd.DataFrame(rows, columns=["display_name", "family", "mode", "score_col"])


def get_high_rep_exon_ids(rep_df):
    return set(
        rep_df.loc[
            pd.to_numeric(rep_df["median_replicate_rho"], errors="coerce") > REPLICATE_RHO_THRESHOLD,
            EXON_COL,
        ]
        .dropna()
        .astype(str)
    )


def filter_source_df_to_high_rep(df, rep_df):
    high_rep_ids = get_high_rep_exon_ids(rep_df)
    return df[df[EXON_COL].astype(str).isin(high_rep_ids)].copy()


def filter_metrics_to_high_rep(metrics_df, rep_df):
    high_rep_ids = get_high_rep_exon_ids(rep_df)
    return metrics_df[metrics_df["exon_group_id"].astype(str).isin(high_rep_ids)].copy()

# ============================================================
# BUILD DATA OBJECTS
# ============================================================

SOURCE_DF = add_mut_plot_columns(SOURCE_DF)
MODEL_DF = build_model_definition_table(SOURCE_DF)

df_none = filter_df(SOURCE_DF, "none")
df_wtpsi = filter_df(SOURCE_DF, "wt_psi")
df_effect = filter_df(SOURCE_DF, "abs_effect")


def build_per_exon_metrics(df_filtered, model_df):
    exon_col = choose_exon_group_col(df_filtered)
    dfg = add_groupings(df_filtered)

    model_df_use = model_df[model_df["score_col"].isin(dfg.columns)].copy()

    rows = []
    group_cols = [exon_col, "region", "mut_plot"]

    for model_row in model_df_use.itertuples(index=False):
        tmp = dfg[group_cols + [EXP_COL, model_row.score_col]].copy()
        tmp = tmp.rename(columns={model_row.score_col: "pred_score"})

        tmp["pred_score"] = pd.to_numeric(tmp["pred_score"], errors="coerce")
        tmp[EXP_COL] = pd.to_numeric(tmp[EXP_COL], errors="coerce")
        tmp = tmp.dropna(subset=[EXP_COL, "pred_score"]).copy()

        for keys, g in tmp.groupby(group_cols, sort=False):
            exon_group_id, region, mut_plot = keys

            rho = safe_spearman_no_warning(
                g[EXP_COL],
                g["pred_score"],
                min_n=MIN_MUTS_PER_EXON_GROUP,
            )

            pred_scaled = g["pred_score"] * PRED_SCALE_FOR_ERROR
            signed_error = g[EXP_COL] - pred_scaled

            rows.append({
                "exon_group_id": exon_group_id,
                "region": region,
                "mut_plot": mut_plot,
                "display_name": model_row.display_name,
                "family": model_row.family,
                "mode": model_row.mode,
                "n_mut": int(len(g)),
                "rho": rho,
                "mae": float(np.abs(signed_error).mean()) if len(g) > 0 else np.nan,
            })

    return pd.DataFrame(rows)


def get_fair_subset(metric_df):
    needed_models = set(metric_df["display_name"].unique())

    fair_groups = (
        metric_df
        .dropna(subset=["rho"])
        .groupby(["exon_group_id", "region", "mut_plot"])["display_name"]
        .agg(lambda x: set(x))
        .reset_index(name="model_set")
    )

    fair_groups = fair_groups[
        fair_groups["model_set"].apply(lambda s: needed_models.issubset(s))
    ].copy()

    return metric_df.merge(
        fair_groups[["exon_group_id", "region", "mut_plot"]],
        on=["exon_group_id", "region", "mut_plot"],
        how="inner",
    )


metrics_none = build_per_exon_metrics(df_none, MODEL_DF)
metrics_wtpsi = build_per_exon_metrics(df_wtpsi, MODEL_DF)
metrics_effect = build_per_exon_metrics(df_effect, MODEL_DF)

# ============================================================
# REPLICATE REPRODUCIBILITY TABLE
# ============================================================

required_rep_cols = [EXON_COL, "psi_r1", "psi_r2", "psi_r3", "se_wt_psi"]
missing_rep_cols = [c for c in required_rep_cols if c not in SOURCE_DF.columns]
if missing_rep_cols:
    raise ValueError(f"Missing required replicate / uncertainty columns:\n{missing_rep_cols}")

tmp_rep = SOURCE_DF.copy()
for c in ["psi_r1", "psi_r2", "psi_r3", "se_wt_psi"]:
    tmp_rep[c] = pd.to_numeric(tmp_rep[c], errors="coerce")

rep_rows = []

for exon_id, g in tmp_rep.groupby(EXON_COL, dropna=False):
    rep12 = safe_spearman_no_warning(g["psi_r1"], g["psi_r2"])
    rep13 = safe_spearman_no_warning(g["psi_r1"], g["psi_r3"])
    rep23 = safe_spearman_no_warning(g["psi_r2"], g["psi_r3"])

    vals = [rep12, rep13, rep23]
    vals = [x for x in vals if pd.notna(x)]

    median_rep_rho = np.median(vals) if len(vals) > 0 else np.nan

    rep_rows.append({
        EXON_COL: exon_id,
        "rep12_rho": rep12,
        "rep13_rho": rep13,
        "rep23_rho": rep23,
        "median_replicate_rho": median_rep_rho,
        "n_mutations_for_replicate_rho": len(g),
    })

rep_df = pd.DataFrame(rep_rows)

wtpsi_df = (
    tmp_rep.groupby(EXON_COL, dropna=False)["se_wt_psi"]
    .median()
    .reset_index()
    .rename(columns={"se_wt_psi": "median_se_wt_psi"})
)

high_rep_exons = get_high_rep_exon_ids(rep_df)

print("\n" + "=" * 100)
print("REPLICATE REPRODUCIBILITY FILTER")
print("=" * 100)
print(f"Threshold: {REPLICATE_FILTER_LABEL}")
print("Exons passing threshold:", len(high_rep_exons))
print("All exons with replicate estimates:", rep_df[EXON_COL].nunique(dropna=True))

rep_df.to_csv(
    os.path.join(OUT_DIR_FINAL, "per_exon_replicate_reproducibility.tsv"),
    sep="\t",
    index=False,
)

# ============================================================
# POOLED PANEL / RESIDUAL DATA BUILDERS
# ============================================================

def collapse_for_pooled_panel(df_filtered, model_df):
    model_cols = [c for c in model_df["score_col"].tolist() if c in df_filtered.columns]

    required_cols = PANEL_COLLAPSE_KEYS + [EXP_COL, "region", "mut_plot"]
    optional_cols = [
        c for c in [
            "ensembl_exon_id",
            "exon_id",
            "variant_id",
            "variant_token",
            "mut",
            "mut_type",
            "wt_psi",
        ]
        if c in df_filtered.columns and c not in required_cols
    ]

    tmp = df_filtered[required_cols + model_cols + optional_cols].copy()

    agg = {
        EXP_COL: "mean",
        "region": "first",
        "mut_plot": "first",
        **{c: "mean" for c in model_cols},
        **{c: "first" for c in optional_cols},
    }

    return tmp.groupby(PANEL_COLLAPSE_KEYS, as_index=False, dropna=False).agg(agg)


def build_pooled_panel_df(df_filtered, model_df):
    df_base = collapse_for_pooled_panel(df_filtered, model_df)
    dfg = add_groupings(df_base)

    model_df_use = model_df[model_df["score_col"].isin(dfg.columns)].copy()

    rows = []
    for model_row in model_df_use.itertuples(index=False):
        tmp = dfg[["region", "mut_plot", EXP_COL, model_row.score_col]].copy()
        tmp = tmp.rename(columns={model_row.score_col: "pred_score"})

        tmp["pred_score"] = pd.to_numeric(tmp["pred_score"], errors="coerce")
        tmp[EXP_COL] = pd.to_numeric(tmp[EXP_COL], errors="coerce")
        tmp = tmp.dropna(subset=[EXP_COL, "pred_score"]).copy()

        for (region, mut_plot), g in tmp.groupby(["region", "mut_plot"], sort=False):
            rho = safe_spearman_no_warning(
                g[EXP_COL],
                g["pred_score"],
                min_n=MIN_MUTS_FOR_POOLED_PANEL,
            )

            pred_scaled = g["pred_score"] * PRED_SCALE_FOR_ERROR
            signed_error = g[EXP_COL] - pred_scaled

            rows.append({
                "region": region,
                "mut_plot": mut_plot,
                "display_name": model_row.display_name,
                "family": model_row.family,
                "mode": model_row.mode,
                "rho": rho,
                "mae": float(np.abs(signed_error).mean()),
                "n_mutations": int(len(g)),
            })

    return pd.DataFrame(rows)


def get_pooled_region_counts(df_filtered, model_df):
    df_base = collapse_for_pooled_panel(df_filtered, model_df)
    dfg = add_groupings(df_base)
    return dfg.query("mut_plot == 'All'").groupby("region").size().to_dict()


def build_pooled_long_for_prediction_errors(df_filtered, model_df):
    exon_col = choose_exon_group_col(df_filtered)

    parts = []
    for model_row in model_df.itertuples(index=False):
        tmp = df_filtered[[exon_col, EXP_COL, model_row.score_col]].copy()
        tmp = tmp.rename(columns={
            exon_col: "exon_group_id",
            model_row.score_col: "pred_score",
        })

        tmp["display_name"] = model_row.display_name
        tmp["family"] = model_row.family
        tmp["mode"] = model_row.mode

        for c in ["ensembl_exon_id", "exon_id", "variant_token", "variant_id", "nt_seq"]:
            if c in df_filtered.columns:
                tmp[c] = df_filtered[c].values

        parts.append(tmp)

    long_df = pd.concat(parts, ignore_index=True)
    long_df["pred_score"] = pd.to_numeric(long_df["pred_score"], errors="coerce")
    long_df[EXP_COL] = pd.to_numeric(long_df[EXP_COL], errors="coerce")
    long_df = long_df.dropna(subset=["pred_score", EXP_COL]).copy()

    long_df["pred_scaled"] = long_df["pred_score"] * PRED_SCALE_FOR_ERROR
    long_df["signed_prediction_error"] = long_df[EXP_COL] - long_df["pred_scaled"]
    long_df["absolute_prediction_error"] = np.abs(long_df["signed_prediction_error"])

    return long_df


def compute_prediction_error_correlation_matrices(df_filtered, model_df):
    long_df = build_pooled_long_for_prediction_errors(df_filtered, model_df)

    idx_cols = ["nt_seq", "variant_token"]
    idx_cols = [c for c in idx_cols if c in long_df.columns]
    if not idx_cols:
        idx_cols = ["exon_group_id"]

    wide_signed = long_df.pivot_table(
        index=idx_cols,
        columns="display_name",
        values="signed_prediction_error",
        aggfunc="first",
        observed=False,
    )

    wide_abs = long_df.pivot_table(
        index=idx_cols,
        columns="display_name",
        values="absolute_prediction_error",
        aggfunc="first",
        observed=False,
    )

    corr_signed = wide_signed.corr(method="pearson", min_periods=50)
    corr_abs = wide_abs.corr(method="pearson", min_periods=50)

    return corr_signed, corr_abs, wide_signed, wide_abs

# ============================================================
# PLOTTING FUNCTIONS
# ============================================================

def plot_panelA_final(metrics_df, title_suffix, metric, x_label, filename, single_row=True):
    fair_df = (
        get_fair_subset(metrics_df)
        .query("region == 'All' and mut_plot == 'All'")
        .copy()
    )

    families = [
        fam for fam in MODEL_FAMILY_ORDER
        if f"{fam} genome" in fair_df["display_name"].unique()
        and f"{fam} minigene" in fair_df["display_name"].unique()
    ]

    if not single_row:
        row_modes = ["genome", "minigene"]

        all_counts = []
        for mode in row_modes:
            for fam in families:
                model_name = f"{fam} {mode}"
                vals = pd.to_numeric(
                    fair_df.loc[fair_df["display_name"] == model_name, metric],
                    errors="coerce",
                ).dropna().values

                counts, _ = (
                    np.histogram(vals, bins=40, range=(-0.55, 1.0))
                    if metric == "rho"
                    else np.histogram(vals, bins=40)
                )
                all_counts.append(counts.max() if len(counts) else 0)

        ymax = int(np.ceil(max(all_counts) * 1.10)) if all_counts else 1
        n_exons = fair_df["exon_group_id"].nunique()

        fig, axes = plt.subplots(
            2,
            len(families),
            figsize=(2.2 * len(families), 3.95),
            dpi=220,
            sharex=True,
            sharey=True,
        )

        for i, mode in enumerate(row_modes):
            for j, fam in enumerate(families):
                ax = axes[i, j]
                model_name = f"{fam} {mode}"

                vals = pd.to_numeric(
                    fair_df.loc[fair_df["display_name"] == model_name, metric],
                    errors="coerce",
                ).dropna().values

                med, q1, q3 = median_iqr(vals)

                hist_kwargs = dict(
                    bins=40,
                    color=PALETTE[model_name],
                    edgecolor="black",
                    linewidth=0.35,
                    alpha=0.85,
                )
                if metric == "rho":
                    hist_kwargs["range"] = (-0.55, 1.0)

                ax.hist(vals, **hist_kwargs)
                ax.axvline(med, color="black", linestyle="--", linewidth=0.9)

                if pd.notna(q1) and pd.notna(q3):
                    ax.axvspan(q1, q3, color="black", alpha=0.08)

                ax.text(
                    0.03,
                    0.97,
                    f"Median {med:.2f}\nIQR [{q1:.2f}, {q3:.2f}]",
                    ha="left",
                    va="top",
                    transform=ax.transAxes,
                    fontsize=7.2,
                )

                if i == 0:
                    ax.set_title(fam, fontsize=9, pad=6)

                if j == len(families) - 1:
                    ax.text(
                        1.08,
                        0.5,
                        mode,
                        rotation=-90,
                        transform=ax.transAxes,
                        ha="left",
                        va="center",
                        fontsize=9,
                    )

                if metric == "rho":
                    ax.set_xlim(-0.55, 1.0)

                ax.set_ylim(0, ymax)
                ax.tick_params(axis="x", pad=1.5)
                ax.tick_params(axis="y", pad=1.5)
                apply_r_panel_theme(ax)

        fig.supylabel("Number of exons", fontsize=9, x=0.008)
        fig.supxlabel(x_label, fontsize=9, y=0.055)

        fig.suptitle(
            f"{title_suffix} (n = {n_exons} exons)",
            y=1.015,
            fontsize=11,
        )

        plt.tight_layout(rect=[0.025, 0.065, 0.98, 0.98], h_pad=0.55, w_pad=0.6)
        save_jpeg(fig, filename)
        plt.show()
        return

    model_order = [
        *(f"{fam} genome" for fam in families),
        *(f"{fam} minigene" for fam in families),
    ]

    all_counts = []
    for model_name in model_order:
        vals = pd.to_numeric(
            fair_df.loc[fair_df["display_name"] == model_name, metric],
            errors="coerce",
        ).dropna().values

        counts, _ = (
            np.histogram(vals, bins=40, range=(-0.55, 1.0))
            if metric == "rho"
            else np.histogram(vals, bins=40)
        )
        all_counts.append(counts.max() if len(counts) else 0)

    ymax = int(np.ceil(max(all_counts) * 1.10)) if all_counts else 1
    n_exons = fair_df["exon_group_id"].nunique()

    fig, axes = plt.subplots(
        1,
        len(model_order),
        figsize=(15.6, 3.35),
        dpi=260,
        sharex=True,
        sharey=True,
    )

    for j, model_name in enumerate(model_order):
        ax = axes[j]

        vals = pd.to_numeric(
            fair_df.loc[fair_df["display_name"] == model_name, metric],
            errors="coerce",
        ).dropna().values

        med, q1, q3 = median_iqr(vals)

        hist_kwargs = dict(
            bins=40,
            color=PALETTE[model_name],
            edgecolor="black",
            linewidth=0.35,
            alpha=0.85,
        )

        if metric == "rho":
            hist_kwargs["range"] = (-0.55, 1.0)

        ax.hist(vals, **hist_kwargs)
        ax.axvline(med, color="black", linestyle="--", linewidth=1.0)

        if pd.notna(q1) and pd.notna(q3):
            ax.axvspan(q1, q3, color="black", alpha=0.08)

        fam = model_name.replace(" genome", "").replace(" minigene", "")

        ax.set_title(fam, fontsize=11, pad=6)

        ax.text(
            0.03,
            0.97,
            f"Median {med:.2f}\nIQR [{q1:.2f}, {q3:.2f}]",
            ha="left",
            va="top",
            transform=ax.transAxes,
            fontsize=10,
        )

        if metric == "rho":
            ax.set_xlim(-0.55, 1.0)
            ax.set_xticks([-0.5, 0.0, 0.5, 1.0])
            ax.set_xticklabels(["-0.5", "0", "0.5", "1"])

        ax.set_ylim(0, ymax)

        ax.tick_params(axis="x", pad=1.5, labelsize=13)
        ax.tick_params(axis="y", pad=1.5, labelsize=13)

        apply_r_panel_theme(ax)

    fig.text(
        0.27,
        0.985,
        "Genome",
        ha="center",
        va="bottom",
        fontsize=13,
        fontweight="bold",
    )

    fig.text(
        0.73,
        0.985,
        "Minigene",
        ha="center",
        va="bottom",
        fontsize=13,
        fontweight="bold",
    )

    fig.supylabel("Number of exons", fontsize=12, x=0.004)
    fig.supxlabel(x_label, fontsize=12, y=0.04)

    fig.suptitle(
        f"{title_suffix} (n = {n_exons} exons)",
        y=1.105,
        fontsize=13,
    )

    plt.tight_layout(rect=[0.02, 0.065, 0.995, 0.93], w_pad=0.40)
    save_jpeg(fig, filename)
    plt.show()


def plot_panelA_v2_replicate_overlay(metrics_df, rep_df, title_suffix, metric, x_label, filename):
    fair_df_all = (
        get_fair_subset(metrics_df)
        .query("region == 'All' and mut_plot == 'All'")
        .copy()
    )

    high_rep_ids = get_high_rep_exon_ids(rep_df)

    fair_df_all["exon_group_id_str"] = fair_df_all["exon_group_id"].astype(str)
    fair_df_high = fair_df_all[fair_df_all["exon_group_id_str"].isin(high_rep_ids)].copy()

    families = [
        fam for fam in MODEL_FAMILY_ORDER
        if f"{fam} genome" in fair_df_all["display_name"].unique()
        and f"{fam} minigene" in fair_df_all["display_name"].unique()
    ]

    row_modes = ["genome", "minigene"]

    all_counts = []
    for mode in row_modes:
        for fam in families:
            model_name = f"{fam} {mode}"

            vals_all = pd.to_numeric(
                fair_df_all.loc[fair_df_all["display_name"] == model_name, metric],
                errors="coerce",
            ).dropna().values

            vals_high = pd.to_numeric(
                fair_df_high.loc[fair_df_high["display_name"] == model_name, metric],
                errors="coerce",
            ).dropna().values

            hist_range = (-0.55, 1.0) if metric == "rho" else None
            counts_all, _ = np.histogram(vals_all, bins=40, range=hist_range)
            counts_high, _ = np.histogram(vals_high, bins=40, range=hist_range)

            all_counts.extend([
                counts_all.max() if len(counts_all) else 0,
                counts_high.max() if len(counts_high) else 0,
            ])

    ymax = int(np.ceil(max(all_counts) * 1.10)) if all_counts else 1

    n_exons_all = fair_df_all["exon_group_id"].nunique()
    n_exons_high = fair_df_high["exon_group_id"].nunique()

    summary_rows = []

    fig, axes = plt.subplots(
        2,
        len(families),
        figsize=(2.2 * len(families), 3.95),
        dpi=220,
        sharex=True,
        sharey=True,
    )

    for i, mode in enumerate(row_modes):
        for j, fam in enumerate(families):
            ax = axes[i, j]
            model_name = f"{fam} {mode}"

            base_color = PALETTE[model_name]
            pale_color = lighten_color(base_color, amount=0.62)

            vals_all = pd.to_numeric(
                fair_df_all.loc[fair_df_all["display_name"] == model_name, metric],
                errors="coerce",
            ).dropna().values

            vals_high = pd.to_numeric(
                fair_df_high.loc[fair_df_high["display_name"] == model_name, metric],
                errors="coerce",
            ).dropna().values

            med_all, q1_all, q3_all = median_iqr(vals_all)
            med_high, q1_high, q3_high = median_iqr(vals_high)

            hist_range = (-0.55, 1.0) if metric == "rho" else None

            ax.hist(
                vals_all,
                bins=40,
                range=hist_range,
                color=pale_color,
                edgecolor="none",
                linewidth=0,
                alpha=0.55,
                label="All constructs",
            )

            ax.hist(
                vals_high,
                bins=40,
                range=hist_range,
                color=base_color,
                edgecolor="black",
                linewidth=0.35,
                alpha=0.88,
                label=REPLICATE_FILTER_LABEL,
            )

            ax.axvline(med_all, color=pale_color, linestyle="--", linewidth=0.85)
            ax.axvline(med_high, color="black", linestyle="--", linewidth=1.05)

            if pd.notna(q1_high) and pd.notna(q3_high):
                ax.axvspan(q1_high, q3_high, color="black", alpha=0.08)

            ax.text(
                0.03,
                0.97,
                f"Median {med_high:.2f}\nIQR [{q1_high:.2f}, {q3_high:.2f}]",
                ha="left",
                va="top",
                transform=ax.transAxes,
                fontsize=7.4,
                color="black",
            )

            ax.text(
                0.03,
                0.73,
                f"All median {med_all:.2f}\nAll IQR [{q1_all:.2f}, {q3_all:.2f}]",
                ha="left",
                va="top",
                transform=ax.transAxes,
                fontsize=6.5,
                color=pale_color,
            )

            summary_rows.append({
                "display_name": model_name,
                "family": fam,
                "mode": mode,
                "metric": metric,
                "n_exons_all": len(vals_all),
                "all_median": med_all,
                "all_q1": q1_all,
                "all_q3": q3_all,
                "n_exons_inter_replicate_experimental_spearman_R_gt_0_8": len(vals_high),
                "filtered_median": med_high,
                "filtered_q1": q1_high,
                "filtered_q3": q3_high,
            })

            if i == 0:
                ax.set_title(fam, fontsize=9, pad=6)

            if j == len(families) - 1:
                ax.text(
                    1.08,
                    0.5,
                    mode,
                    rotation=-90,
                    transform=ax.transAxes,
                    ha="left",
                    va="center",
                    fontsize=9,
                )

            if metric == "rho":
                ax.set_xlim(-0.55, 1.0)

            ax.set_ylim(0, ymax)
            ax.tick_params(axis="x", pad=1.5)
            ax.tick_params(axis="y", pad=1.5)
            apply_r_panel_theme(ax)

    handles, labels = axes[0, 0].get_legend_handles_labels()
    fig.legend(
        handles,
        labels,
        loc="upper center",
        bbox_to_anchor=(0.5, 0.995),
        ncol=2,
        frameon=False,
        fontsize=7.5,
    )

    fig.supylabel("Number of exons", fontsize=9, x=0.008)
    fig.supxlabel(x_label, fontsize=9, y=0.055)

    fig.suptitle(
        f"{title_suffix}\nall n = {n_exons_all}; filtered n = {n_exons_high}",
        y=1.065,
        fontsize=11,
    )

    plt.tight_layout(rect=[0.025, 0.065, 0.98, 0.93], h_pad=0.55, w_pad=0.6)
    save_jpeg(fig, filename)
    plt.show()

    summary_df = pd.DataFrame(summary_rows)

    print("\n" + "=" * 100)
    print("SUPPLEMENTARY PANEL A v2 MEDIANS / IQR")
    print("=" * 100)
    display(summary_df)

    summary_df.to_csv(
        os.path.join(
            OUT_DIR_FINAL,
            filename.replace(".jpg", "_median_iqr_summary.tsv").replace(".jpeg", "_median_iqr_summary.tsv"),
        ),
        sep="\t",
        index=False,
    )

    return summary_df


def plot_supp_between_model_correlations(metrics_df, title_suffix, filename, restrict_to_high_rep=False, rep_df=None):
    GENOME_COLOR = "steelblue"
    DIAG_COLOR = "#999999"
    MINIGENE_COLOR = "#b23a48"

    panel = get_fair_subset(metrics_df).query("region == 'All' and mut_plot == 'All'").copy()

    if restrict_to_high_rep:
        if rep_df is None:
            raise ValueError("rep_df is required when restrict_to_high_rep=True")

        high_rep_ids = get_high_rep_exon_ids(rep_df)
        panel = panel[panel["exon_group_id"].astype(str).isin(high_rep_ids)].copy()

    wide = panel.pivot_table(
        index="exon_group_id",
        columns=["family", "mode"],
        values="rho",
        aggfunc="first",
        observed=False,
    )

    families = [
        fam for fam in MODEL_FAMILY_ORDER
        if (fam, "genome") in wide.columns and (fam, "minigene") in wide.columns
    ]

    need_cols = [(fam, mode) for fam in families for mode in ["genome", "minigene"]]
    wide = wide.dropna(subset=need_cols)

    n = len(families)

    fig, axes = plt.subplots(
        n,
        n,
        figsize=(1.9 * n, 1.32 * n),
        dpi=240,
        sharex=True,
        sharey=True,
    )

    tick_vals = [0, 0.25, 0.50, 0.75]
    tick_labs = ["0", "0.25", "0.5", "0.75"]

    for i, fam_y in enumerate(families):
        for j, fam_x in enumerate(families):
            ax = axes[i, j]

            if i == j:
                x = wide[(fam_x, "genome")]
                y = wide[(fam_y, "minigene")]
                point_color = DIAG_COLOR
                r, _, _ = safe_pearson_stats(x, y)
                label_txt = f"{fam_x}\nR = {r:.3f}"
            elif i > j:
                x = wide[(fam_x, "genome")]
                y = wide[(fam_y, "genome")]
                point_color = GENOME_COLOR
                r, _, _ = safe_pearson_stats(x, y)
                label_txt = f"R = {r:.3f}"
            else:
                x = wide[(fam_x, "minigene")]
                y = wide[(fam_y, "minigene")]
                point_color = MINIGENE_COLOR
                r, _, _ = safe_pearson_stats(x, y)
                label_txt = f"R = {r:.3f}"

            ax.scatter(x, y, s=5, alpha=0.40, color=point_color, linewidths=0)
            ax.text(0.04, 0.94, label_txt, ha="left", va="top", transform=ax.transAxes, fontsize=7.0)

            ax.set_xlim(-0.05, 1.0)
            ax.set_ylim(-0.05, 1.0)
            ax.set_xticks(tick_vals)
            ax.set_yticks(tick_vals)
            ax.set_xticklabels(tick_labs)
            ax.set_yticklabels(tick_labs)
            apply_r_panel_theme(ax)

            if i == n - 1:
                ax.set_xlabel(fam_x, fontsize=9)
                ax.tick_params(axis="x", labelbottom=True, pad=1.5)
            else:
                ax.tick_params(axis="x", labelbottom=False)

            if j == 0:
                ax.set_ylabel(fam_y, fontsize=9)
                ax.tick_params(axis="y", labelleft=True, pad=1.5)
            else:
                ax.tick_params(axis="y", labelleft=False)

    fig.suptitle(f"{title_suffix} (n = {wide.shape[0]} exons)", fontsize=11, y=0.985)
    plt.tight_layout(rect=[0, 0, 1, 0.94], h_pad=0.35, w_pad=0.45)

    save_jpeg(fig, filename)
    plt.show()


def plot_panelB_pooled_region_mutation(df_filtered, model_df, metric, filename):
    pooled = build_pooled_panel_df(df_filtered, model_df)

    model_order = [
        f"{fam} {mode}"
        for fam in MODEL_FAMILY_ORDER
        for mode in ["genome", "minigene"]
        if f"{fam} {mode}" in pooled["display_name"].unique()
    ]

    region_order_present = [
        r for r in REGION_ORDER
        if r in pooled["region"].astype(str).unique()
    ]

    region_counts = get_pooled_region_counts(df_filtered, model_df)

    y_label = {
        "rho": "Pooled Spearman R\nacross all mutations",
        "mae": "Pooled MAE\n|observed − predicted|",
    }[metric]

    fig, axes = plt.subplots(
        1,
        len(region_order_present),
        figsize=(3.2 * len(region_order_present), 5.0),
        dpi=240,
        sharey=True,
    )

    if len(region_order_present) == 1:
        axes = [axes]

    bar_width = 0.9 / max(1, len(model_order))
    x_centers = np.arange(len(MUT_PLOT_ORDER))

    for ax, region in zip(axes, region_order_present):
        reg_df = pooled[pooled["region"].astype(str) == region].copy()
        n_region = region_counts.get(region, np.nan)

        title_txt = region if pd.isna(n_region) else f"{region}\n({int(n_region):,} mutations)"
        ax.set_title(title_txt, fontsize=13, pad=10)

        for k, model_name in enumerate(model_order):
            sub = reg_df[reg_df["display_name"] == model_name].copy()
            sub["mut_plot"] = sub["mut_plot"].astype(str)

            sub = (
                sub
                .set_index("mut_plot")
                .reindex(MUT_PLOT_ORDER)
                .reset_index()
            )

            xpos = x_centers - 0.45 + bar_width / 2 + k * bar_width
            y = sub[metric].values.astype(float)

            ax.bar(
                xpos,
                y,
                width=bar_width,
                color=PALETTE[model_name],
                edgecolor="none",
                alpha=0.95,
            )

        ax.set_xticks(x_centers)

        ax.set_xticklabels(
            MUT_PLOT_ORDER,
            fontsize=15,
            rotation=45,
            ha="right",
        )

        ax.tick_params(axis="y", labelsize=14)

        if metric == "rho":
            ax.set_ylim(0, 1.0)

        ax.spines["top"].set_visible(False)
        ax.spines["right"].set_visible(False)

    axes[0].set_ylabel(y_label, fontsize=17)

    fig.supxlabel("Mutation", fontsize=17, y=0.02)

    plt.suptitle(
        f"Fig. 4B. Pooled Spearman R",
        y=1.04,
        fontsize=18,
    )

    plt.tight_layout(rect=[0, 0.04, 1, 1], w_pad=0.70)
    save_jpeg(fig, filename)
    plt.show()

    return pooled


red_white_purple = LinearSegmentedColormap.from_list(
    "red_white_purple",
    ["#B2182B", "#FFFFFF", "#762A83"],
)


def plot_residual_heatmap_final(corr_df, title, filename):
    ordered_models = [
        f"{fam} {mode}"
        for fam in MODEL_FAMILY_ORDER
        for mode in ["genome", "minigene"]
        if f"{fam} {mode}" in corr_df.index and f"{fam} {mode}" in corr_df.columns
    ]

    corr_df = corr_df.loc[ordered_models, ordered_models]
    mat = corr_df.values.astype(float)

    fig, ax = plt.subplots(figsize=(6.4, 3.85), dpi=260)

    im = ax.imshow(mat, cmap=red_white_purple, vmin=-1, vmax=1, aspect="auto")

    ax.set_xticks(np.arange(len(ordered_models)))
    ax.set_yticks(np.arange(len(ordered_models)))
    ax.set_xticklabels(ordered_models, rotation=45, ha="right", fontsize=8)
    ax.set_yticklabels(ordered_models, fontsize=8)

    for i in range(mat.shape[0]):
        for j in range(mat.shape[1]):
            val = mat[i, j]
            ax.text(j, i, "NA" if pd.isna(val) else f"{val:.2f}", ha="center", va="center", fontsize=7)

    for spine in ax.spines.values():
        spine.set_visible(False)

    ax.tick_params(axis="both", length=0, pad=1.5)

    cbar = fig.colorbar(im, ax=ax, shrink=0.70, pad=0.02)
    cbar.set_label("Pearson R", fontsize=8.5)
    cbar.ax.tick_params(labelsize=7)

    ax.set_title(title, fontsize=10.5, pad=10)

    plt.tight_layout()

    save_jpeg(fig, filename)
    plt.show()

# ============================================================
# SUMMARY TABLES
# ============================================================

panelA_df = (
    get_fair_subset(metrics_none)
    .query("region == 'All' and mut_plot == 'All'")
    .copy()
)

model_perf = (
    panelA_df[
        ["display_name", "family", "mode", "exon_group_id", "rho", "mae"]
    ]
    .rename(columns={
        "exon_group_id": EXON_COL,
        "rho": "model_rho",
        "mae": "model_mae",
    })
    .copy()
)

compare_df = (
    model_perf
    .merge(rep_df, on=EXON_COL, how="inner")
    .merge(wtpsi_df, on=EXON_COL, how="left")
)

summary_rows = []

for model_name in MODEL_ORDER:
    sub = compare_df[compare_df["display_name"].eq(model_name)].copy()

    rep_corr = safe_spearman_no_warning(
        sub["median_replicate_rho"],
        sub["model_rho"],
    )

    wtpsi_corr = safe_spearman_no_warning(
        sub["median_se_wt_psi"],
        sub["model_rho"],
    )

    summary_rows.append({
        "display_name": model_name,
        "replicate_reproducibility_corr": rep_corr,
        "wtpsi_se_corr": wtpsi_corr,
        "n_exons": sub[EXON_COL].nunique(),
    })

summary_df = pd.DataFrame(summary_rows)

print("\n" + "=" * 100)
print("MODEL VS EXPERIMENTAL QUALITY CORRELATIONS")
print("=" * 100)
display(summary_df)

summary_df.to_csv(
    os.path.join(OUT_DIR_FINAL, "model_vs_experimental_quality_correlations.tsv"),
    sep="\t",
    index=False,
)

# ============================================================
# MAIN FIGURE 4
# ============================================================

plot_panelA_final(
    metrics_df=metrics_none,
    title_suffix="Fig. 4A. Per-exon Spearman R distributions",
    metric="rho",
    x_label="Per-exon Spearman R",
    filename="fig_4a_per_exon_spearman_single_row.jpg",
    single_row=True,
)

panelB_pooled = plot_panelB_pooled_region_mutation(
    df_filtered=df_none,
    model_df=MODEL_DF,
    metric="rho",
    filename="fig_4b_pooled_spearman_by_region_and_mutation_class.jpg",
)

# ============================================================
# SUPPLEMENTARY FIGURES
# ============================================================

plot_panelA_final(
    metrics_df=metrics_none,
    title_suffix="Supplementary Fig. Per-exon Spearman R distributions — two-row version",
    metric="rho",
    x_label="Per-exon Spearman R",
    filename="supp_per_exon_spearman_two_row.jpg",
    single_row=False,
)

panelA_v2_summary = plot_panelA_v2_replicate_overlay(
    metrics_df=metrics_none,
    rep_df=rep_df,
    title_suffix="Supplementary Fig. Per-exon Spearman R distributions with reproducible-construct overlay",
    metric="rho",
    x_label="Per-exon Spearman R",
    filename="supp_per_exon_spearman_inter_replicate_experimental_spearman_R_gt_0_8_overlay.jpg",
)

plot_supp_between_model_correlations(
    metrics_df=metrics_none,
    title_suffix="Supplementary Fig. Agreement of exon-level predictive performance across model families — no filter",
    filename="supp_between_model_performance_correlations.jpg",
)

plot_supp_between_model_correlations(
    metrics_df=metrics_none,
    title_suffix=f"Supplementary Fig. Agreement across model families — {REPLICATE_FILTER_LABEL} only",
    filename="supp_between_model_performance_correlations_inter_replicate_experimental_spearman_R_gt_0_8.jpg",
    restrict_to_high_rep=True,
    rep_df=rep_df,
)

plot_panelA_final(
    metrics_df=metrics_wtpsi,
    title_suffix=f"Supplementary Fig. WT PSI {WT_PSI_MIN}–{WT_PSI_MAX} per-exon MAE distributions",
    metric="mae",
    x_label="Per-exon MAE",
    filename="supp_wt_psi_20_80_per_exon_mae.jpg",
    single_row=True,
)

metrics_wtpsi_high_rep = filter_metrics_to_high_rep(metrics_wtpsi, rep_df)

plot_panelA_final(
    metrics_df=metrics_wtpsi_high_rep,
    title_suffix=(
        f"Supplementary Fig. WT PSI {WT_PSI_MIN}–{WT_PSI_MAX}, "
        f"{REPLICATE_FILTER_LABEL}, per-exon MAE distributions"
    ),
    metric="mae",
    x_label="Per-exon MAE",
    filename="supp_wt_psi_20_80_inter_replicate_spearman_R_gt_0_8_per_exon_mae.jpg",
    single_row=True,
)

corr_signed_none, corr_abs_none, wide_signed_none, wide_abs_none = compute_prediction_error_correlation_matrices(
    df_none,
    MODEL_DF,
)

plot_residual_heatmap_final(
    corr_signed_none,
    "Supplementary Fig. Signed residual correlations across models — no filter",
    "supp_signed_residual_correlations.jpg",
)

df_none_high_rep = filter_source_df_to_high_rep(df_none, rep_df)

corr_signed_high_rep, corr_abs_high_rep, wide_signed_high_rep, wide_abs_high_rep = compute_prediction_error_correlation_matrices(
    df_none_high_rep,
    MODEL_DF,
)

plot_residual_heatmap_final(
    corr_signed_high_rep,
    f"Supplementary Fig. Signed residual correlations across models — {REPLICATE_FILTER_LABEL}",
    "supp_signed_residual_correlations_inter_replicate_experimental_spearman_R_gt_0_8.jpg",
)

# ============================================================
# STANDALONE LEGEND
# ============================================================

legend_order = [
    "SpliceTransformer genome",
    "SpliceTransformer minigene",
    "AlphaGenome genome",
    "AlphaGenome minigene",
    "Pangolin genome",
    "Pangolin minigene",
    "SpliceAI genome",
    "SpliceAI minigene",
]

handles = [
    Patch(
        facecolor=PALETTE[name],
        edgecolor="none",
        label=name,
    )
    for name in legend_order
]

fig, ax = plt.subplots(figsize=(3.0, 3.9), dpi=300)

ax.legend(
    handles=handles,
    loc="center",
    frameon=False,
    fontsize=15,
    ncol=1,
    handlelength=1.4,
    handleheight=0.8,
    labelspacing=0.45,
    borderpad=0.15,
    handletextpad=0.5,
)

ax.axis("off")

save_jpeg(
    fig,
    "standalone_vertical_legend_single_column_compact.jpg",
)

plt.show()

print("\nFinal JPEG figure folder:")
print(OUT_DIR_FINAL)
```
